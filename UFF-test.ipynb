```python
import mbuild as mb 
import foyer
```

```python
zeolite = mb.load("vpi5-bonded.mol2")
zeolite.visualize().show()
```

```python
print(set([p.name for p in zeolite.particles()]))
```

```python
uff = foyer.forcefield.Forcefield()
uff.loadFile("uff.xml")
```

```python
typed_zeolite = uff.apply(zeolite, assert_bond_params=False, assert_angle_params=False, assert_dihedral_params=False)
```

```python
all_types = set([atom.type for atom in typed_zeolite.atoms])

print(all_types)
```
